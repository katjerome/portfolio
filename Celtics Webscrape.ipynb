```python
# Import library to open URLs

import urllib.request

# Import the BeautifulSoup library so we can parse HTML and XML documents
from bs4 import BeautifulSoup
# Web-scrape URL
url = "https://en.wikipedia.org/wiki/2023%E2%80%9324_Boston_Celtics_season"
# Open the Url and put the html into the page variable
page = urllib.request.urlopen(url)

# Parse the HTML from the NBA url into the BeautifulSoup parse tree format
soup = BeautifulSoup(page, 'html.parser')
```

```python
# find all table tags
nba_table = soup.find_all('table')

# Print the type & length of nba_table
print(type(nba_table))
print(len(nba_table))

# Iterate thru nba_table and print each table entry

#Commented out for space purposes

#for table in nba_table:
    #print(f"Sortable Table: {table}\n")
```

```python
# Find all table with class_="wikitable"
nba_table_sort = soup.find(class_="toccolours")

# Print length of nba_table_sort
print(len(nba_table_sort))

# Iterate thru nba_table_sort
for table in nba_table_sort:
    print(f"Sortable Tables: {table}\n")
```

```python
for row in nba_table_sort.find_all('tr'):
    cells = row.find_all('td')
    print(f"Number of cells in this row: {len(cells)}")
    for cell in cells:
        print(cell.get_text(strip=True))

    
```

```python

# Create a list for each column of data
A=[]
B=[]
C=[]
D=[]
E=[]
F=[]
G=[]

# Ignore the headers and loop through the rows to get data for each Subtraction

for row in nba_table_sort.find_all('tr'):
    cells = row.find_all('td')
    if len(cells) == 7:
            A.append(cells[0].get_text(strip=True))
            B.append(cells[1].get_text(strip=True))
            C.append(cells[2].get_text(strip=True))
            D.append(cells[3].get_text(strip=True))
            E.append(cells[4].get_text(strip=True))
            F.append(cells[5].get_text(strip=True))
            G.append(cells[6].get_text(strip=True))
            
print("Pos.:", A)
print("No.:", B)
print("Name:", C)
print("Height:", D)
print("Weight:", E)
print("DOB:", F)
print("From:", G)
```

# Celtics Roster during the 23/24 Season 

```python
# Import pandas library

import pandas as pd

# Create a Dataframe using column  'A' to start, 

df=pd.DataFrame(A,columns=['Pos'])
df['No']=B
df['Name']=C
df['Height']=D
df['Weight']=E
df['DOB']=F
df['From']=G
# Display the DataFrame
df
```

```python
# Create the csv file for the data set

df.to_csv('CSC221-webscrape-data.csv', index=False)

# Print file created message
print("CSC221-webscrape-data.csv created!!")
print("All done :-)")
```

```python
#For our plotting we will be taking an average of the player's weight in each position and plotting it
# in a bar graph
#print(df)
dfSORT = df.sort_values(by=['Weight'], ascending=True)
#print(dfSORT)

PLAYER_POS = np.array(dfSORT.Pos)

WEIGHTS = np.array(dfSORT.Weight)

# Unique positions for color coding
unique_positions = dfSORT['Pos'].unique()

# Color dictionary for positions
color_dict = {pos: plt.cm.tab20(i) for i, pos in enumerate(unique_positions)}

# Assign colors based on positions
colors = [color_dict[pos] for pos in PLAYER_POS]

X_VALUES = list(PLAYER_POS)
Y_VALUES = list(WEIGHTS)

# Plot
plt.bar(X_VALUES, Y_VALUES, color=colors)
plt.title('Celtic Weights', fontsize=12, fontweight='bold')
plt.xlabel('Player Positions')
plt.ylabel('Weights')

# Create custom legend
handles = [plt.Line2D([0], [0], color=color_dict[pos], lw=4) for pos in unique_positions]
labels = unique_positions
plt.legend(handles, labels, title="Positions")

plt.show()
```
